```python
import requests
import folium

def wkwkkw(ip_address):
    url = f"https://ipinfo.io/{ip_address}/json"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        loc = data.get("loc", "0,0")  
        latitude, longitude = map(float, loc.split(",")) if "," in loc else (0, 0)
        
        details = {
            "IP si kocak": ip_address,
            "Latitude": latitude,
            "Longitude": longitude,
            "Kota": data.get("city", "Kota tak ditemukeun"),
            "Wilayah": data.get("region", "Wilayah tak ditemukeun"),
            "Negara": data.get("country", "Negara tak ditemukeun"),
            "Kode Pos": data.get("postal", "Kode pos tak ditemukeun"),
            "Timezone": data.get("timezone", "Timezone tak ditemukeun"),
            "ISP": data.get("org", "ISP tak ditemukeun")
        }

        print("Data IP si kocak:")
        for key, value in details.items():
            print(f"{key}: {value}")

        fig = folium.Figure(width=600, height=400)
        map_ip = folium.Map(location=[latitude, longitude], zoom_start=35)
        map_ip.add_to(fig)
        
        folium.Marker([latitude, longitude], popup=f"IP: {ip_address}\nKota: {details['Kota']}").add_to(map_ip)

        return fig
    else:
        print(f"Error mas: {response.status_code}")

ip = "175.176.92.90"
wkwkkw(ip)
```
